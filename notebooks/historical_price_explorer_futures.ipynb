```python
# === APPLY SYMBOL TRANSLATION FOR FUTURES ===
print("=== APPLYING SYMBOL TRANSLATION ===\n")

# Define the translation mapping
symbol_mapping = {
    'TY': 'TYU5 Comdty',
    'FV': 'FVU5 Comdty',
    'TU': 'TUU5 Comdty',
    'US': 'USU5 Comdty',
    'RX': 'RXU5 Comdty'
}

# Create a copy for translation
translated_df = combined_df.copy()

# Show current symbols
print("Current symbols in dataframe:")
print(translated_df['symbol'].unique())

# Apply translation
print("\nApplying translations...")
translated_df['symbol'] = translated_df['symbol'].map(symbol_mapping)

# Check if all symbols were translated
if translated_df['symbol'].isna().any():
    print("\n WARNING: Some symbols could not be translated:")
    unmapped = combined_df[translated_df['symbol'].isna()]['symbol'].unique()
    print(f"Unmapped symbols: {unmapped}")
else:
    print("\n All symbols successfully translated!")

# Show translated symbols
print("\nTranslated symbols:")
print(translated_df['symbol'].unique())

# Display sample of translated data
print("\nSample of translated data:")
print(translated_df.head(10))
```

```python
# === PREPARE FUTURES DATA FOR DATABASE INSERTION ===
print("=== PREPARING FUTURES DATA FOR DATABASE ===\n")

# Start with translated dataframe
insert_df = translated_df.copy()

# 1. Normalize symbol spacing (just in case)
print("1. Normalizing symbol spacing...")
insert_df['symbol'] = insert_df['symbol'].astype(str).str.strip()
insert_df['symbol'] = insert_df['symbol'].str.replace(r'\s+', ' ', regex=True)

# 2. Remove any duplicates
print("\n2. Checking for duplicates...")
before_dedup = len(insert_df)
insert_df = insert_df.drop_duplicates(subset=['symbol', 'timestamp'], keep='last')
after_dedup = len(insert_df)
print(f"   Removed {before_dedup - after_dedup} duplicate rows")

# 3. Format timestamp to include seconds
print("\n3. Formatting timestamps...")
insert_df['timestamp'] = insert_df['timestamp'] + ':00'
print(f"   Sample timestamp: {insert_df['timestamp'].iloc[0]}")

# 4. Convert price to string
print("\n4. Converting price to string...")
insert_df['price'] = insert_df['price'].astype(str)

# 5. Final verification
print("\n=== FINAL DATA READY FOR INSERTION ===")
print(f"Total rows: {len(insert_df)}")
print(f"Data types: {dict(insert_df.dtypes)}")
print(f"Unique symbols: {insert_df['symbol'].unique()}")
print(f"\nSample data:")
print(insert_df.head())

# Save for insertion
futures_insert_df = insert_df.copy()
print("\n Data saved to 'futures_insert_df' variable for insertion")
```

```python
# === INSERT FUTURES DATA INTO DATABASE ===
import sqlite3

print("=== INSERTING FUTURES DATA INTO DATABASE ===\n")

# Connect to database
conn = sqlite3.connect('../trades.db')
cursor = conn.cursor()

try:
    # Check current state
    cursor.execute("SELECT COUNT(*) FROM pricing")
    initial_count = cursor.fetchone()[0]
    print(f"Current rows in pricing table: {initial_count}")
    
    # Check specifically for our futures symbols before insertion
    futures_symbols = ['TYU5 Comdty', 'FVU5 Comdty', 'TUU5 Comdty', 'USU5 Comdty', 'RXU5 Comdty']
    print("\nChecking for existing futures data (our specific symbols):")
    for symbol in futures_symbols:
        cursor.execute("SELECT COUNT(*) FROM pricing WHERE symbol = ? AND price_type = 'close'", (symbol,))
        count = cursor.fetchone()[0]
        print(f"   {symbol}: {count} existing rows")
    
    # Convert dataframe to list of tuples for insertion
    records = futures_insert_df[['symbol', 'price_type', 'price', 'timestamp']].values.tolist()
    print(f"\nInserting {len(records)} futures records...")
    
    # Insert in batches
    batch_size = 500
    total_inserted = 0
    
    for i in range(0, len(records), batch_size):
        batch = records[i:i+batch_size]
        
        # Use INSERT OR REPLACE to handle conflicts
        cursor.executemany("""
            INSERT OR REPLACE INTO pricing (symbol, price_type, price, timestamp)
            VALUES (?, ?, ?, ?)
        """, batch)
        
        total_inserted += len(batch)
        print(f"   Batch {i//batch_size + 1}: Inserted {len(batch)} rows")
    
    # Commit the transaction
    conn.commit()
    print(f"\n Successfully inserted {total_inserted} futures records")
    
    # Get final count
    cursor.execute("SELECT COUNT(*) FROM pricing")
    final_count = cursor.fetchone()[0]
    print(f"\nFinal row count in pricing table: {final_count}")
    print(f"Net new rows added: {final_count - initial_count}")
    
    # Show sample of our specific futures data
    print("\nSample of newly inserted futures data (our symbols only):")
    cursor.execute("""
        SELECT symbol, price_type, price, timestamp 
        FROM pricing 
        WHERE symbol IN ('TYU5 Comdty', 'FVU5 Comdty', 'TUU5 Comdty', 'USU5 Comdty', 'RXU5 Comdty')
        AND price_type = 'close'
        ORDER BY timestamp DESC 
        LIMIT 10
    """)
    
    for row in cursor.fetchall():
        print(f"   {row[0]:<20} | {row[1]:<10} | {row[2]:<10} | {row[3]}")
    
    # Summary by our specific futures symbols
    print("\nFutures data summary by symbol (our symbols only):")
    cursor.execute("""
        SELECT symbol, COUNT(*) as count
        FROM pricing 
        WHERE symbol IN ('TYU5 Comdty', 'FVU5 Comdty', 'TUU5 Comdty', 'USU5 Comdty', 'RXU5 Comdty')
        AND price_type = 'close'
        GROUP BY symbol
        ORDER BY symbol
    """)
    
    for row in cursor.fetchall():
        print(f"   {row[0]}: {row[1]} rows")
        
except Exception as e:
    print(f" Error during insertion: {e}")
    conn.rollback()
    
finally:
    conn.close()
    print("\n Database connection closed")
```

```python
# === DIAGNOSTIC: WHAT'S REALLY GOING ON ===
import sqlite3

print("=== DIAGNOSTIC CHECK ===\n")

conn = sqlite3.connect('../trades.db')
cursor = conn.cursor()

try:
    # 1. Check the actual table schema
    print("1. PRICING TABLE SCHEMA:")
    cursor.execute("PRAGMA table_info(pricing)")
    schema = cursor.fetchall()
    for col in schema:
        print(f"   {col}")
    
    # 2. Check if there's a primary key or unique constraint
    print("\n2. TABLE CONSTRAINTS:")
    cursor.execute("SELECT sql FROM sqlite_master WHERE type='table' AND name='pricing'")
    create_statement = cursor.fetchone()
    if create_statement:
        print(f"   {create_statement[0]}")
    
    # 3. Check what's actually in the database for our futures symbols
    print("\n3. EXISTING DATA FOR OUR FUTURES SYMBOLS:")
    futures_symbols = ['TYU5 Comdty', 'FVU5 Comdty', 'TUU5 Comdty', 'USU5 Comdty', 'RXU5 Comdty']
    
    for symbol in futures_symbols:
        cursor.execute("""
            SELECT symbol, price_type, price, timestamp 
            FROM pricing 
            WHERE symbol = ?
            ORDER BY timestamp
        """, (symbol,))
        
        rows = cursor.fetchall()
        print(f"\n   {symbol}: {len(rows)} rows")
        if rows:
            for row in rows[:3]:  # Show first 3
                print(f"      {row}")
                
    # 4. Check all price_types in the database
    print("\n4. ALL PRICE_TYPES IN DATABASE:")
    cursor.execute("SELECT DISTINCT price_type FROM pricing")
    price_types = cursor.fetchall()
    for pt in price_types:
        print(f"   '{pt[0]}'")
        
    # 5. Check our dataframe's unique combinations
    print("\n5. OUR DATAFRAME'S UNIQUE (symbol, price_type) COMBINATIONS:")
    unique_combos = futures_insert_df[['symbol', 'price_type']].drop_duplicates()
    print(f"   Total unique combinations: {len(unique_combos)}")
    for _, row in unique_combos.iterrows():
        print(f"   {row['symbol']} | {row['price_type']}")
        
    # 6. Find the exact conflict
    print("\n6. FINDING THE EXACT CONFLICT:")
    for _, row in futures_insert_df.iterrows():
        cursor.execute("""
            SELECT COUNT(*) 
            FROM pricing 
            WHERE symbol = ? AND price_type = ?
        """, (row['symbol'], row['price_type']))
        
        count = cursor.fetchone()[0]
        if count > 0:
            print(f"   CONFLICT: {row['symbol']} | {row['price_type']} already exists")
            break
            
except Exception as e:
    print(f"Error: {e}")
    
finally:
    conn.close()

print("\n=== END DIAGNOSTIC ===")
```

```python
# === SAFELY MODIFY PRICING TABLE TO INCLUDE TIMESTAMP IN PRIMARY KEY ===
import sqlite3

print("=== MODIFYING PRICING TABLE STRUCTURE (SAFELY) ===\n")

conn = sqlite3.connect('../trades.db')
cursor = conn.cursor()

try:
    # 1. Create a backup table with ALL existing data
    print("1. Creating backup of existing pricing data...")
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS pricing_backup AS 
        SELECT * FROM pricing
    """)
    
    cursor.execute("SELECT COUNT(*) FROM pricing_backup")
    backup_count = cursor.fetchone()[0]
    print(f"    Backed up {backup_count} rows to pricing_backup table")
    
    # 2. Rename the original table
    print("\n2. Renaming original table...")
    cursor.execute("ALTER TABLE pricing RENAME TO pricing_old")
    print("    Renamed pricing  pricing_old")
    
    # 3. Create new pricing table with timestamp in primary key
    print("\n3. Creating new pricing table with better structure...")
    cursor.execute("""
        CREATE TABLE pricing (
            symbol TEXT NOT NULL,
            price_type TEXT NOT NULL,
            price REAL,
            timestamp TEXT NOT NULL,
            PRIMARY KEY (symbol, price_type, timestamp)
        )
    """)
    print("    Created new table with PRIMARY KEY (symbol, price_type, timestamp)")
    
    # 4. Copy all data from old table to new
    print("\n4. Copying all data to new table...")
    cursor.execute("""
        INSERT INTO pricing (symbol, price_type, price, timestamp)
        SELECT symbol, price_type, price, timestamp FROM pricing_old
    """)
    copied = cursor.rowcount
    print(f"    Copied {copied} rows")
    
    # 5. Verify data integrity
    print("\n5. Verifying data integrity...")
    cursor.execute("SELECT COUNT(*) FROM pricing")
    new_count = cursor.fetchone()[0]
    
    cursor.execute("SELECT COUNT(*) FROM pricing_old")
    old_count = cursor.fetchone()[0]
    
    if new_count == old_count:
        print(f"    Data verified: {new_count} rows in new table matches {old_count} in old")
        
        # 6. Drop the old table
        print("\n6. Cleaning up...")
        cursor.execute("DROP TABLE pricing_old")
        print("    Dropped pricing_old table")
    else:
        print(f"    Data mismatch! New: {new_count}, Old: {old_count}")
        print("   Rolling back changes...")
        raise Exception("Data integrity check failed")
    
    # Commit all changes
    conn.commit()
    print("\n Successfully modified table structure!")
    
    # 7. Now insert our futures data
    print("\n7. Now inserting futures historical data...")
    records = futures_insert_df[['symbol', 'price_type', 'price', 'timestamp']].values.tolist()
    
    cursor.executemany("""
        INSERT OR IGNORE INTO pricing (symbol, price_type, price, timestamp)
        VALUES (?, ?, ?, ?)
    """, records)
    
    inserted = cursor.rowcount
    conn.commit()
    print(f"    Inserted {inserted} new futures records")
    
    # 8. Final summary
    print("\n8. Final summary:")
    cursor.execute("SELECT COUNT(*) FROM pricing")
    final_total = cursor.fetchone()[0]
    print(f"   Total rows in pricing table: {final_total}")
    
    # Show futures data
    for symbol in ['TYU5 Comdty', 'FVU5 Comdty', 'TUU5 Comdty', 'USU5 Comdty', 'RXU5 Comdty']:
        cursor.execute("""
            SELECT COUNT(*) 
            FROM pricing 
            WHERE symbol = ? AND price_type = 'close'
        """, (symbol,))
        count = cursor.fetchone()[0]
        print(f"   {symbol}: {count} close prices")
        
except Exception as e:
    print(f"\n Error occurred: {e}")
    print("Attempting to rollback...")
    conn.rollback()
    
    # Try to restore original state
    try:
        cursor.execute("DROP TABLE IF EXISTS pricing")
        cursor.execute("ALTER TABLE pricing_old RENAME TO pricing")
        print(" Restored original table")
    except:
        print(" Could not restore - check pricing_backup table for your data")
    
finally:
    conn.close()

print("\n=== DONE ===")
print("Your pricing table now supports multiple prices per symbol!")
print("The backup is in 'pricing_backup' table if needed")
```
