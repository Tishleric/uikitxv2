```python
# === CLEAN DATA PREPARATION FOR DATABASE INSERTION ===
print("=== PREPARING DATA FOR DATABASE INSERTION ===\n")

# Start with our cleaned dataframe (non-numeric prices already removed)
insert_df = combined_df.copy()
print(f"Starting with {len(insert_df)} rows")

# 0. First, normalize symbols by replacing multiple spaces with single space
print("\n0. Normalizing symbol spacing...")
# Convert symbol to string and strip leading/trailing whitespace
insert_df['symbol'] = insert_df['symbol'].astype(str).str.strip()
# Replace multiple spaces with single space
insert_df['symbol'] = insert_df['symbol'].str.replace(r'\s+', ' ', regex=True)
print("   ✓ Replaced multiple spaces with single spaces in symbols")

# 1. Remove invalid symbols (0, 0.0, empty, etc.)
print("\n1. Checking for and removing invalid symbols (0, 0.0, empty, etc.)...")
# Find problematic symbols
invalid_symbols = insert_df[
    (insert_df['symbol'] == '0') | 
    (insert_df['symbol'] == '0.0') |
    (insert_df['symbol'] == '') |
    (insert_df['symbol'].isna()) |
    (insert_df['symbol'].str.match(r'^[\d\.]+$', na=False))  # Any pure number
]

if len(invalid_symbols) > 0:
    print(f"   Found {len(invalid_symbols)} rows with invalid symbols:")
    print(f"   Symbol values: {invalid_symbols['symbol'].value_counts().to_dict()}")
    print(f"   Sample of invalid rows:")
    print(invalid_symbols[['symbol', 'price', 'timestamp']].head())
    
    # Remove these rows
    insert_df = insert_df[~insert_df.index.isin(invalid_symbols.index)]
    print(f"   Removed {len(invalid_symbols)} rows with invalid symbols")
    print(f"   Rows remaining: {len(insert_df)}")

# 2. Remove exact duplicates (where ALL columns are the same)
print("\n2. Removing exact duplicates...")
before_dedup = len(insert_df)
insert_df = insert_df.drop_duplicates()
after_dedup = len(insert_df)
print(f"   Removed {before_dedup - after_dedup} exact duplicate rows")

# 3. Check for symbol-timestamp duplicates with different prices
print("\n3. Checking for symbol-timestamp duplicates with different prices...")
dup_check = insert_df.groupby(['symbol', 'timestamp']).agg({
    'price': ['count', 'nunique', lambda x: list(x.unique())]
}).reset_index()
dup_check.columns = ['symbol', 'timestamp', 'count', 'unique_prices', 'price_list']
conflicts = dup_check[dup_check['unique_prices'] > 1]

if len(conflicts) > 0:
    print(f"   WARNING: Found {len(conflicts)} symbol-timestamp pairs with DIFFERENT prices:")
    for _, row in conflicts.head(10).iterrows():
        print(f"   Symbol: {row['symbol']}, Time: {row['timestamp']}, Prices: {row['price_list']}")
else:
    print("   ✓ No price conflicts found")

# 4. Remove remaining duplicates, keeping last
print("\n4. Removing remaining duplicates (keeping last occurrence)...")
before_final = len(insert_df)
insert_df = insert_df.drop_duplicates(subset=['symbol', 'timestamp'], keep='last')
after_final = len(insert_df)
print(f"   Removed {before_final - after_final} duplicate rows")

# 5. Format timestamp to include seconds
print("\n5. Formatting timestamps...")
insert_df['timestamp'] = insert_df['timestamp'] + ':00'
print(f"   Sample timestamp: {insert_df['timestamp'].iloc[0]}")

# 6. Convert price to string
print("\n6. Converting price to string...")
insert_df['price'] = insert_df['price'].astype(str)

# 7. Final verification
print("\n=== FINAL DATA READY FOR INSERTION ===")
print(f"Total rows: {len(insert_df)}")
print(f"Data types: {dict(insert_df.dtypes)}")
print(f"\nSample data with normalized symbols:")
print(insert_df.head())

# Double-check for any remaining weird symbols
weird_symbols = insert_df[insert_df['symbol'].str.match(r'^[\d\.]+$', na=False)]
if len(weird_symbols) > 0:
    print(f"\nERROR: Still found {len(weird_symbols)} rows with numeric-only symbols!")
    print("This should not happen - please investigate further")
else:
    print(f"\n✓ All symbols are valid (no numeric-only symbols found)")
```

```python
# === DATABASE INSERTION ===
print("=== INSERTING DATA INTO DATABASE ===\n")

# Connect to database
conn = sqlite3.connect('../trades.db')
cursor = conn.cursor()

try:
    # Get initial count
    cursor.execute("SELECT COUNT(*) FROM pricing")
    initial_count = cursor.fetchone()[0]
    print(f"Initial row count in pricing table: {initial_count}")
    
    # Convert dataframe to list of tuples for insertion
    records = insert_df[['symbol', 'price_type', 'price', 'timestamp']].values.tolist()
    
    # Insert in batches for better performance
    batch_size = 500
    total_inserted = 0
    
    for i in range(0, len(records), batch_size):
        batch = records[i:i+batch_size]
        
        # Use INSERT OR REPLACE to handle conflicts with existing data
        cursor.executemany("""
            INSERT OR REPLACE INTO pricing (symbol, price_type, price, timestamp)
            VALUES (?, ?, ?, ?)
        """, batch)
        
        total_inserted += len(batch)
        print(f"   Inserted batch {i//batch_size + 1}: {len(batch)} rows (Total: {total_inserted}/{len(records)})")
    
    # Commit the transaction
    conn.commit()
    print(f"\n✓ Successfully inserted/updated {total_inserted} records")
    
    # Get final count
    cursor.execute("SELECT COUNT(*) FROM pricing")
    final_count = cursor.fetchone()[0]
    print(f"\nFinal row count in pricing table: {final_count}")
    print(f"Net new rows added: {final_count - initial_count}")
    
    # Show sample of inserted data
    print("\nSample of data in database after insertion:")
    cursor.execute("SELECT * FROM pricing WHERE price_type = 'close' ORDER BY timestamp DESC LIMIT 5")
    for row in cursor.fetchall():
        print(f"   {row}")
        
except Exception as e:
    print(f"❌ Error during insertion: {e}")
    conn.rollback()
    
finally:
    conn.close()
    print("\n✓ Database connection closed")
```
